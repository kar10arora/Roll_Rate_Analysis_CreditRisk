```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
roll_rate = pd.read_excel("D:\Credit Risk Model\Roll Rate Analysis\Roll rate.xlsx")
roll_rate
```

```python
roll_rate['Reporting Month'] = pd.to_datetime(roll_rate['Reporting Month'])
roll_rate['Origination date'] = pd.to_datetime(roll_rate['Origination date'])
roll_rate
```

# MOB (Month on Book) Method

```python
roll_rate['MOB'] = roll_rate.apply(lambda x : (x['Reporting Month'].year - x['Origination date'].year)*12 
                                   + x['Reporting Month'].month - x['Origination date'].month - 1, axis = 1 )
roll_rate
```

```python
roll_rate['MOB'].min() , roll_rate['MOB'].max()
```

```python
for i in range(144,240,12):
    print(i)
```

```python
def capping(data):
    if data['MOB']<156:
        return 144
    elif data['MOB']<168:
        return 156
    elif data['MOB']<180:
        return 168
    elif data['MOB']<192:
        return 180
    elif data['MOB']<204:
        return 192
    elif data['MOB']<=238:
        return 204
```

```python
roll_rate['MOB_cap'] = roll_rate.apply(capping, axis=1)
roll_rate['MOB_cap'].value_counts()
```

```python
roll_rate_pivot = roll_rate.pivot_table(index = 'Loan Id',
                             values = 'Reporting Month',
                             aggfunc = 'count')
roll_rate_pivot
```

```python
roll_rate_check = pd.merge(roll_rate , roll_rate_pivot , on = 'Loan Id' , how = 'left')
roll_rate_check
```

```python
roll_rate_check['check1'] = np.where(roll_rate_check['Reporting Month_y']<12 , 1,0)
roll_rate_check
```

```python
roll_rate_cleaned = roll_rate_check[~(roll_rate_check['check1']==1)]
roll_rate_cleaned
```
